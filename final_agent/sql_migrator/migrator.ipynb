```python
import pandas as pd
from prisma import Prisma


df = pd.read_csv("2025-Jan.csv")


df["year"] = 2025
df["month"] = "Jan"


db = Prisma()

async def load_data():
    await db.connect()

    
    records = df.to_dict(orient="records")

    
    await db.Jan2025.create_many(data=records)

    await db.disconnect()
```
